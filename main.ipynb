```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import warnings
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error

warnings.simplefilter(action='ignore', category=FutureWarning)
```

# Problem approach
There are many disasters in the world to consider which contribute to humanitarian need. We decided to look at the consequence of disasters around the world: refugees. The approach is to make a simple time series prediction and build the model up iteratively such that it can incoporate different disasters in it's predictions. 

The datasets which we used for this challenge comprised of the following:
* World Bank Open Data : obtaining data on the human capital index (HCI)
* IATI: For obtaining data on the spendings of financial resources for humanitarian purposes
* FTS: Similar as AITI, which was helpful for reference material
* UNHCR: Extract monthly data of refugees from a collection of countries

**First we load and clean up data**

```python
# Load data
rdf = pd.read_csv('data/refugee_monthly.csv')
rdf.columns = ['Country', 'Origin', 'Year', 'Month', 'Value']

# Clean up data
col = 'Value'
rdf[col] = pd.to_numeric(rdf[col], errors='coerce')
rdf[col] = rdf[col].fillna(2)

month_mapping = {'January':1, 'February':2, 'April':3, 'May':4, 'July':5, 'August':6,'September':7, 'October':8, 'November':9, 'June':10, 'March':11, 'December': 12}
rdf.Month = rdf.Month.map(month_mapping)

# Create initial frame
init_frame = rdf.groupby(["Year", "Month", "Country"]).size().reset_index().drop([0], axis=1)
```

## Here we try to find the "hotspots" of refugees. Basically we are trying to find out which countries produce the most number of refugees (globally) and potentially use them as features as well.

```python
# Finding global heavy hitters
rdf = pd.read_csv('data/refugee_monthly.csv')
rdf.columns = ['Country', 'Origin', 'Year', 'Month', 'Value']

col = 'Value'
rdf[col] = pd.to_numeric(rdf[col], errors='coerce')
rdf[col] = rdf[col].fillna(rdf[col].median())


ref_counts = rdf.groupby(['Origin'], as_index=False)['Value'].sum()
sort_ref = ref_counts.sort_values('Value', ascending=False)
global_hh = sort_ref[['Origin', 'Value']].reset_index(drop=True)
plt.plot(global_hh.Value)

# Find local heavy hitters (specific to a certain hotspots)
def find_ref_hotspots(ref_data, country, high_hitters=10):
    country_data = ref_data[ref_data.Country==country]
    ref_counts = country_data.groupby(['Origin'], as_index=False)['Value'].sum()
    sort_ref = ref_counts.sort_values('Value', ascending=False)
    return sort_ref["Origin"][:high_hitters].tolist()
    
# Retrieve heavy hitters of country
print(find_ref_hotspots(rdf, "Turkey"))
```

## Here we extract historical data and are able to use it for predictive analyses by means of simple time-series forecast

```python
def add_historical_data(iiframe, ref_data, look_back=5):
    iref_data = ref_data.copy(deep=True)
    iref_data = iref_data.groupby(["Year", "Month", "Country"]).sum().reset_index()
    iframe = iiframe.copy(deep = True)
    
    X = iref_data 
    for i in range(1, look_back):
        X['t-'+str(i)] = iref_data.Value.shift(i)
        
    X = X[look_back-1:]
    history_cols = X.columns[3:]
    iframe[history_cols] = X[history_cols]
    iframe['y'] = X.Value
    iframe = iframe.dropna()
    iframe = iframe.drop(['Value'], axis=1)
    return iframe


add_historical_data(init_frame, rdf)
```

```python
# Add heavy hitters current data (determined later in the notebook)
# Human Capital Index
heavy_hitters = ['Syrian Arab Rep.', 'Afghanistan', 'Iraq', 'Serbia and Kosovo: S/RES/1244 (1999)', 'China', 'Russian Federation', 'Iran (Islamic Rep. of)', 'Pakistan', 'Somalia', 'Nigeria', 'Eritrea', 'Turkey', 'Mexico', 'Albania', 'Sri Lanka', 'India', 'Bangladesh', 'Various/unknown', 'Dem. Rep. of the Congo', 'Georgia', 'Algeria', 'El Salvador', 'Haiti', 'Ukraine', 'Armenia', 'Stateless', 'Colombia', 'Guinea', 'Guatemala', 'Bosnia and Herzegovina']

def add_hhcurr_data(iiframe, rdata):
    iframe = iiframe.copy(deep=True)
    for i, hh in enumerate(heavy_hitters):
        cfilter = rdata[rdata.Origin == hh]
        cfilter_next = cfilter[['Year','Month','Value']].groupby(['Year','Month'], as_index=False).sum()
        iframe['{}_curr'.format(hh)] = cfilter_next['Value']
        iframe.columns = [''.join(col) for col in iframe.columns]
    return iframe

add_hhcurr_data(init_frame, rdf)
```

## This part allows us to look at the inform indices in order to consider local data in a country where people are leaving (origin of the refugee)

```python
# Add INFORM data of heavy hitters
inform_data = pd.read_csv('data/inform_index.csv')
# inform_data = inform_data[inform_data.columns[0:5]]

def add_inform_data(iiframe, iidata):
    idata = iidata.copy(deep=True)
    idata = idata[idata['country'].isin(heavy_hitters)]
    iframe = iiframe.copy(deep=True)
    iframe = pd.merge(iframe, idata, on=['Year'])
    iframe = pd.pivot_table(iframe, index=['Year', 'Month', 'Country'], columns='country', values=iframe.columns[3:]).reset_index()
    return iframe

add_inform_data(init_frame, inform_data)
# inform_data.columns
```

## Here we build and test the model to see how it performs over all the countries and determine an average score. We train a XGBoost model with the help of sklearn library.

```python
inform_indices = ['Year', 'country', 'Agriculture Stress Index Probability', 'Conflict Barometer National Power','People affected by drought (absolute) - raw', 'Human Develpment Index']

all_countries = rdf['Country'].unique()
all_errors = []

for c in all_countries:
    temp_frame = init_frame[init_frame['Country'] == c].reset_index(drop=True)
    
#     temp_frame = add_inform_data(temp_frame, inform_data[inform_indices])
#     temp_frame = add_hhcurr_data(temp_frame, rdf)
    temp_frame = add_historical_data(temp_frame, rdf)
    
    for col in temp_frame.columns[5:]:
        temp_frame[col] = (temp_frame[col]-temp_frame[col].mean())/temp_frame[col].std()
        

    X = temp_frame[temp_frame.columns[3:-1]].as_matrix()
    y = temp_frame[temp_frame.columns[-1]].as_matrix()
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=False, test_size=3)
    rgr = XGBRegressor()
    rgr.fit(X_train, y_train.tolist())
    preds = rgr.predict(X_test)
    error = mean_absolute_error(preds, y_test)/(np.ptp(y))
    all_errors.append(error)

    print('error on {}: {}'.format(c, error))
print('Mean error {}'.format(np.mean(all_errors)))
```

# After the prediction of the prices we try to estimate the cost of humanitarian aid and allocate funds effectively. Our final output is the total cost in a certain country and the fraction of which should be allocated to different sectors (food, emergency, health, etc...)

```python
cost_data = pd.read_csv("data/funds.csv", parse_dates=['date'])
ref_data = pd.read_csv("data/refugee_monthly.csv")

col = 'Value'
ref_data[col] = pd.to_numeric(ref_data[col], errors='coerce')
ref_data[col] = ref_data[col].fillna(2)

month_mapping = {'January':1, 'February':2, 'April':3, 'May':4, 'July':5, 'August':6,'September':7, 'October':8, 'November':9, 'June':10, 'March':11, 'December': 12}
ref_data.Month = ref_data.Month.map(month_mapping)

ref_data_agg = ref_data.groupby(['Year', 'Month', 'Country / territory of asylum/residence'], as_index=False)["Value"].sum()
ref_data_agg
```

```python
# Aggregate data based on how it is distributed over different categories (which we computed with a different script).
year_agg = cost_data.groupby([cost_data.date.dt.year, 'recipient', 'category']).sum().reset_index()
summer = year_agg.groupby(['date', 'recipient'])['transaction_value'].transform('sum')
year_agg['distribution'] = year_agg['transaction_value']/summer
year_agg
```

```python
# Match up with country and date
year_agg = year_agg.rename(columns={'recipient': 'country', 'date': 'year'})
result_data = ref_data_agg.rename(columns={'Country / territory of asylum/residence': 'country', 'Year': 'year', "Value": 'refugees'})

# Manually correct for unmatching countries
resultslocs = result_data['country'].copy(deep=True)
resultslocs[resultslocs == 'USA (INS/DHS)'] = 'United States' 
resultslocs[resultslocs == 'USA (EOIR)'] = 'United States' 
resultslocs[resultslocs == 'Rep. of Korea'] = 'Korea, Republic Of' 
resultslocs[resultslocs == 'United Kingdom of Great Britain and Northern Ireland'] = 'United Kingdom' 
result_data['country'] = resultslocs

final_costs = pd.merge(result_data,year_agg, on=['year', 'country'])
transactions = final_costs.groupby(['year', 'Month', 'country', 'refugees'])['transaction_value'].sum()
final_costs = pd.pivot_table(final_costs, index=['year', 'Month', 'country', 'refugees'], columns='category', values='distribution', aggfunc='sum', fill_value=0).reset_index()
final_costs['expense'] = transactions.tolist()
final_costs.to_csv("cost_distribution_v3.csv", index=False)
final_costs
```

```python
# Predicting next values 
inform_indices = ['Year', 'country', 'Agriculture Stress Index Probability', 'Conflict Barometer National Power','People affected by drought (absolute) - raw', 'Human Develpment Index']

all_countries = rdf['Country'].unique()

for c in all_countries:
    temp_frame = init_frame[init_frame['Country'] == c].reset_index(drop=True)
    
    # Options for adding different data sources
#     temp_frame = add_inform_data(temp_frame, inform_data[inform_indices])
#     temp_frame = add_hhcurr_data(temp_frame, rdf)
    temp_frame = add_historical_data(temp_frame, rdf)

    X = temp_frame[temp_frame.columns[3:-1]].as_matrix()
    y = temp_frame[temp_frame.columns[-1]].as_matrix()
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=False, test_size=1)
    rgr = XGBRegressor()
    rgr.fit(X_train, y_train.tolist())
    preds = rgr.predict(X_test)
    print("{} : {}".format(c, preds))
```
